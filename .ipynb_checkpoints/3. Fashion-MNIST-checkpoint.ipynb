# Logistic Regression: Fashion-MNIST

## 1. Diferencias entre SGDClassifier() y LogisticRegression() 

__LogisticRegression__

La clase LogisticRegression implementa una regresión logistica regularizada, permitiendo utilizar los algoritmos de entrenamiento one vs rest (multi_class = 'ovr') y multinomial cuando se establece el parámetro multi_class como 'multinomial' (esto es similar a softmax logistic regression). Cuando se utiliza 'ovr' el problema se toma como binario para cada clase (etiqueta) del set de datos, cuando es 'multinomial' la optimización del modelo se hace a través de la distribución de probabilidad completa, incluso si se trata de una clasificación binaria.
Esta clase se utiliza para implementar una regresión logística pura, es decir, únicamente sirve para realizar clasificaciones binarias o multiclase. Utilizar la clase ``LogisticRegression()`` es más eficiente cuando se trata de sets de datos relativamente pequeños, pues el tiempo de entrenamiento es un factor importante a tomar encuenta, sobre todo si se trabaja en equipos de bajas especificaciones (como en el caso del equipo de trabajo). 

__SGDClassifier__

El estimador ``SGDClassifier`` implementa un modelo lineal regularizado utilizando el gradiente descendiente estocástico (SGD) como medio de aprendizaje, el ajuste del modelo puede ser controlado mediante el parámetro _loss_, que por default es SVM (supported vector machine). La clase SGDClassifier posibilita realizar clasificaciones binarias y multiclase utilizando como parámetro loss = 'log', la clasificación logistica se realiza a través del clasificador probabilístico (que decide a que clase pertenece cada entrada a través de la probabilidad). Utilizar este estimador fácilita el trabajo con grandes sets de datos, pues al ser estocástico solo se utiliza una pequeña parte de los datos para minimizar el error y ajustar adecuadamente el modelo (limitado a las itereaciones máximas "max_iter").

Los métodos de ambas clases son similares, los más relevantes para este proyecto son:
* ``fit()``: Ajusta el modelo de acuerdo a los datos de entrenamiento.
* ``predict()``: Predice las clases para las muestras dadas.
* ``predict_proba()``: Estima las probabilidades de pertenencia a cada clase.
* ``score()``: regresa la precisión media de los datos seleccionados para el test y las clases reales.

En este caso al final se tienen 60 mil filas de datos con 784 columnas (representando 60 mil imágenes de 28 por 28 píxeles), que para los equipos con los que cuentan los miembros del grupo de trabajo resulta ser computacionalmente pesado. Por lo anteriormente mencionado se optó por utilizar el estimador ``SGDClassifier()`` ya que este se implementa cuando se tienen sets de datos largos y posee los mismos métodos que la clase ``LogisticRegresion()``.

## 2. Aplicación de SGDClassifier para el dataset Fashion-MNIST

### 2.1. Cargando las librerías necesarias

Para este caso el dataset Fashion-MNIST se obtiene de la biblioteca TensorFlow, importando la librería keras. También se cargaron las bibliotecas numpy y matplotlib.

```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
import matplotlib.pyplot as plt
```

La versión de tensor flow utilizada es la siguiente:

### 2.2. Cargando el dataset Fashion-MNIST

Los datos se cargaron como un módulo, el cual se nombró como _"fashion_data"_.

```python
fashion_data = keras.datasets.fashion_mnist
```

```python
type(fashion_data)
```

### 2.3. Obtención de los datos para entrenamiento y test

Los datos para entrenamiento y el testeo se obtuvieron a partir del módulo _"fashion_data"_ y se nombraron de la siguiente manera:

``X_train``: Features de entrenamiento

``X_test``: Features para el test

Los target para el entrenamiento ``y_train`` tienen una longitud (cantidad de datos/filas) de 60,000 correspondiendo a las etiquetas (0-9) que definen a cada imágen.

```python
len(y_train)
```

Los features para el test tienen la forma (10000, 28, 28), que indica un total de 10000 imágenes de 28 por 28 pixeles.

#### 2.4.1. Graficando los datos como imágenes

Los datos deben procesarse antes de introducirlos al modelo de regresión logistica, para una primera visualización de estos se presenta la imágen 4 del set de features para el entrenamiento, este es un vestido, cuya etiqueta como string es "Dress" y 3 como entero. Los datos se encuentran en un rango de 0 a 255, que representan las tonalidades.

```python
plt.figure()
plt.imshow(X_train[3]) # Para graficar la imágen
plt.colorbar() # Insertar la barra de color (0-255)
plt.grid(False) # Ignorando el grid
plt.show() 
```

Ya que los valores se encuentran en un rango de 0 a 255, estos se reescaloron entre 0 y 1, dividiendo las matrices de features entre 255.

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

Se grafican las primeras 30 imágenes en escala de grises (cmap binary) para visualizar las imágenes ya reescaladas entre 0 y 1, como etiqueta en el eje x se presentan las etiquetas representativas de cada tipo de prenda de vestir.

```python
plt.figure(figsize=(10,10))
for i in range(30):
    plt.subplot(6,5,i+1) # 30 celdas (6 x 5), con i+1 representando la posición de la nueva imágen.
    plt.xticks([]) # Se omiten las coordenadas de los ejes
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i], cmap=plt.cm.binary) # Con el cmap se establece el mapa de colores, en este caso es binario (escala de grises)
    plt.xlabel(class_names[y_train[i]]) # Se insertan las etiquetas de los nombres de cada prenda
plt.show()
```

Ya que el tipo de datos de los target son enteros no es necesario realizar alguna conversión (como en el caso de que sean strings).

```python
type(y_train[0])
```

### 2.5. Convirtiendo la matriz 3d a 2d

Para introducir el set de datos de entranamiento utilizando la librería sklearn se necesita que estos se encuentren en matrices de dos dimenciones, ya que se tienen matrices de tres dimenciones a continuación se presenta la conversión de 3D a 2D, a través de la función ``reshape()`` de numpy (se utiliza una tupla para introducir la nueva forma del arreglo), donde se tienen __n__ filas representadas con la función ``len(X)`` y 784 columnas, que representan los pixeles de las imágenes (28 x 28).

```python
X_train = X_train.reshape((len(X_train), 784))
X_test = X_test.reshape((len(X_test), 784))
```

```python
type(X_train)
```

### 2.6. Implementación del modelo con Sklearn

#### 2.6.1. Modelo SGDClassifier con logistic regression

Para el modelo de Logistic Regression se utilizó la librería sklearn, importanto el método ``SGDClassifier()`` (Stochastic Gradient Descent) debido al gran número de datos.

```python
from sklearn.linear_model import SGDClassifier 
```

El modelo fue nombrado como _"model_SGD"_, donde se asignó el método ``SGDClassifier()``, con la siguiente configuración:
* loss = 'log', que indica que se utilizará la función de pérdida "Logistic Regression"
* random_state = 42, semilla utilizado comúnmente para asegurar los mismos resultados en cada corrida
* max_iter = 1500, número máximo de iteraciones, valor por default de 1000

```python
model_SGD = SGDClassifier(loss = 'log', random_state=42, max_iter=1500)
```

Se utiliza el método ``fit`` para ajustar el modelo con los datos de entrenamiento _"X_train"_ (features) y _"y_train"_ (targets).

```python
model_SGD.fit(X = X_train, y = y_train)
```

#### 2.6.2. Predicción en base al modelo generado y los datos seleccionados para el test

Para realizar las predicciones se utiliza el método ``predict``, introduciendo la matriz de datos seleccionados para el test del módelo _"X_train"_, y guardándo los resultados en la variable _"y_predict"_.

```python
y_predict = model_SGD.predict(X_test)
```

Mostrando los arreglos que contienen las clases determinadas con el modelo para cada una de las imágenes del test.

```python
y_predict
```

```python
y_test
```

### 2.7. Evaluación del modelo

Como primera instancia se obtuvo la precisión media del modelo utilizando el método ``score``, la cual fue de 0.8363, a pesar de ser un valor mayor al 80% y que a primera instancia podría parecer completamente aceptable es importante obtener más indicadores para evaluar correctamente el modelo.

```python
model_score = model_SGD.score(X_test, y_test)
```

```python
print('Precisión media:', model_score)
```

#### 2.7.1. Matriz de confusión



Fuente: https://www.ecured.cu/Matrices_de_confusi%C3%B3n

La matriz de confusión es una herramienta utilizada para evaluar el desempeño de algoritmos de clasificación, en esta se presenta la cantidad de datos de acuerdo a como fueron clasificados en la predicción, estos pueden ser:
* Verdaderos positivos: datos que fueron clasificados de manera correcta, presenta positivo en los datos reales y las predicciones.
* Verdaderos negativos: datos que fueron clasificados de manera correcta, presenta negativo en los datos reales y las predicciones.
* Falsos positivos: datos clasificados como positivos en la predicción pero que realmente son negativos.
* Falsos negativos: datos clasificados como negativos en la predicción pero realmente son positivos.

Se utilizó el módulo sklearn.metrics para importar el método ``confusion_matrix`` y obtener la matriz de confusión para las predicciones y los valores reales.

```python
from sklearn.metrics import confusion_matrix
```

Para este caso, donde se tienen 10 clases diferentes, los verdaderos positivos se encuentran en la diagonal principal, el resto de los valores son los que fueron clasificados erroneamente, por ejemplo, en [0, 0] se muestran los verdaderos positivos para la clase 0, en [0, 1] están las imágenes clasificados como clase 1, pero en realidad son clase 0.
Los falsos negativos se encuentran en las horizontales, y los falsos positivos se encuentran en las verticales.

```python
confusion_matrix(y_test, y_predict)
```

Mostrando la matriz de confusión de los datos reales se tiene que para cada clase hay 1000 imágenes (haciendo un total de 10000).

```python
confusion_matrix(y_test, y_test)
```

#### 2.7.2. Indicadores a partir de la matriz de confusión

Para continuar con la evaluación del modelo se utilizó el módulo sklearn.metrics, importando los métodos precision_score, recall_score y f1_score.

```python
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score
```

Para todos la configuración es la siguiente:
* Se introducen las matrices de las clasificaciones reales y las clasificaciones predichas, _"y_test"_ y _"y_predict"_ respectivamente.
* average = None, debe establecerse como None o de lo contrario solo funcionará para clasificación binaria.

##### Precision score

El indicador "precision score" brinda la proporción de los casos predichos como positivos que fueron correctos.

$$precision = \dfrac{TP}{TP+FP}$$
<br/>

```python
precision = precision_score(y_test, y_predict, average = None)
```

```python
print('Precisión para cada clase'.center(60, '-'))
for i in range(len(precision)):
    print(f'Clase {i}: {precision[i].round(3)}')
```

##### Recall score

El indicador "recall score" brinda la proporción de los casos positivos que fueron correctamente identificados (TP / Positivos reales), también se le conoce como razón de verdaderos positivos.

$$recall = \dfrac{TP}{TP+FN}$$

Si se comparan los valores obtenidos con ``recall_score`` con la matriz de confusión puede observarse que representa el porcentaje de imágenes que fueron clasificadas como verdaderos positivos en base al total de verdaderos positivos reales para cada clase.

```python
recall = recall_score(y_test, y_predict, average = None)
```

```python
print('Recall para cada clase'.center(60, '-'))
for i in range(len(recall)):
    print(f'Clase {i}: {recall[i].round(3)}')
```

##### f1 score

El indicador "f1 score" proporciona la media armónica entre la razón de verdaderos positivos (recall) y la precisión, este indicador es más apto para ser utilizado para obtener el promedio de las relaciones y tazas, pues se igualan los pesos de los datos.

$$F1 = \dfrac{2}{\dfrac{1}{precision}+\dfrac{1}{recall}} = 2\times\dfrac{precision\times recall}{precision+recall} = \dfrac{TP}{TP+\dfrac{FN+FP}{2}} $$

```python
f1 = f1_score(y_test, y_predict, average = None)
```

```python
print('f1 para cada clase'.center(60, '-'))
for i in range(len(f1)):
    print(f'Clase {i}: {f1[i].round(3)}')
```

### 2.8. Resultados

De acuerdo con los resultados obtenidos, el menor porcentaje de aciertos se obtuvo para la clase "Shirt (6 entero)", mientras que el mayor porcentaje de aciertos se obtuvo para la clase "Trouser (1 entero)". 

```python
plt.figure(figsize=(14,8))  
eje = np.arange(10)

plt.bar(eje - 0.25, precision*100, color = 'b', width = 0.20,  label='Precision')
plt.bar(eje, recall*100, color = 'g', width = 0.20,  label='Recall')
plt.bar(eje + 0.25, f1*100, color = 'r', width = 0.20,  label='f1')

plt.xticks(ticks = eje, labels = class_names)
plt.xlabel('Indicador', fontsize = 13)
plt.ylabel('Score %', fontsize = 13)
plt.title('Resultados de la evaluación del modelo', fontsize = 16)
plt.legend()

plt.show()
```

Para todas las clases se tuvieron resultados aceptables, mayores a 70%, exceptuando a la clase "Shirt" cuyo f1 score fue de 0.595, estos resultados podrían mejorarse utilizando otros métodos de clasificación, como por ejemplo las redes neuronales, como en el caso presentado en la web _tensorflow.org_ (2021): "__Clasificacion Basica: Predecir una imagen de moda__".

#### 2.8.1. Comparando con otro modelos

El modelo implementado se comparó con el modelo de red neuronal presentado en la web tensorflow.org (2021): "Clasificacion Basica: Predecir una imagen de moda", en donde se obtuvo un Test accuracy de __0.88__, mientras que para este caso, utilizando SGDClassifier se obtuvo un test accuracy de __0.84__.

Uno de los métodos más importantes y útiles con los que cuenta el modelo utilizado SGDClassifier (y que también se encuentra en LogisticRegression) es ``predict_proba``, que retorna las probabilidades estimadas de que cada imágen pertenezca a alguna de las 10 clases.

```python
proba_predict = model_SGD.predict_proba(X_test)
```

Replicando parte de la metodología utilizada en Tensor Flow (2021), se graficarán las imágenes utilizadas para el test en conjunto con una gráfica de barras indicando las probabilidades estimadas de pertenencia para cada clase. Las siguientes funciones fueron tomadas del sitio oficial de __Tensor Flow__ y modificadas a conveniencia para aplicarlas a este proyecto: https://www.tensorflow.org/tutorials/keras/classification.

```python
def plot_image(i, proba_array, true_label, img):
    """
    i: indice de la matriz de datos
    proba_array: arreglo con las probabilidades de predicción de las clases para cada imágen
    true_label: matriz con los valores reales de las clases para cada imágen
    img: matriz de imágenes (features)
    """
    proba_array, true_label, img = proba_array, true_label[i], img[i] # Con i se extrae fila por fila de datos
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

    plt.imshow(img.reshape(28,28), cmap=plt.cm.binary)

    predicted_label = np.argmax(proba_array) # Se extrae el indice donde la probabilidad es mayor, este coincide con la clase
    if predicted_label == true_label:
        color = 'blue' # Para el nombre, azul si se determinó correctamente la clase
    else:
        color = 'red' # Rojo si no se determinó correctamente la clase

    plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label], # Extrae el nombre de la clase de acuerdo a lo predicho
                                100*np.max(proba_array), # Extrae la probabilidad mayor y la convierte en porcentaje
                                class_names[true_label]), # Extrae el nombre de la clase de la matriz class_names
                                color=color) # Se define el color de la impresión

def plot_value_array(i, proba_array, true_label):
    """
    i: indice de la matriz de datos (para usar fácilmente con ciclo for)
    predictios_array: arreglo con las probabilidades de predicción de las clases para cada imágen
    true_label: matriz con los valores reales de las clases para cada imágen
    """
    proba_array, true_label = proba_array, true_label[i]
    plt.grid(False)
    plt.xticks(range(10)) # Se indica que habrá espacio para las 10 clases, se comienza en 0 y termoina en 9
    plt.yticks([])
    thisplot = plt.bar(range(10), proba_array, color="#777777") # Se establece el graficado de las probabilidades de la predicción para la imágen
    plt.ylim([0, 1]) # al no ser mayor a 1 se delimita el eje y en 1 como máximo (1 == 100% de coincidencia con la clase)
    predicted_label = np.argmax(proba_array) # Se extrae la clase con mayor probabilidad (numero entero)

    thisplot[predicted_label].set_color('red') # El predicho lo configura como rojo
    thisplot[true_label].set_color('blue') # El real lo configura como azul
```

A continuación se pone a prueba la función modificada, se grafica la primera imágen (i=0), el color azul del texto en la imágen indica que la clase se determinó correctamente (el color rojo indicaría lo contrario), y en conjunto el texto posee la siguiente forma:

* {Clase predicha} {Probabilidad mayor} {(Clase real del dataset)}

La gráfica de barras de la derecha indica en color azul la clase real, y con rojo la clase predicha, si no hay una barra de color rojo quiere decir que se predijo correctamente, el resto de barras grises (color="#777777") son las demás probabilidades encontradas en el arreglo de probabilidades "_proba_predict_".

```python
i = 0
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, proba_predict[i], y_test, X_test)
plt.subplot(1,2,2)
plot_value_array(i, proba_predict[i],  y_test)
plt.show()
```

```python
num_rows = 5
num_cols = 3
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
    plt.subplot(num_rows, 2*num_cols, 2*i+1)
    plot_image(i, proba_predict[i], y_test, X_test)
    plt.subplot(num_rows, 2*num_cols, 2*i+2)
    plot_value_array(i, proba_predict[i],  y_test)
plt.tight_layout()
plt.show()
```

Comparando el modelo implementado con el modelo de Tensor Flow (2021), se tienen algunas diferencias notables, pues se observa que las probabilidades de algunas imágenes de pertenecer a su clase son mayores con el modelo de red neuronal. También puede observarse que para ambos casos la imagen cuando i = 12 no logra clasificarse adecuadamente, siendo en nuestro modelo de clase "Bag" con un 44% y para la red neuronal "Sandal" con 58%.
